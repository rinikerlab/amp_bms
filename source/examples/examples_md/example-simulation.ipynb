```python
import os
import torch # for determining device
from md.Simulator import Simulator
```

```python
# define some constants and identify device
kSimulationName = "example-simulation"
kPdbInput = os.path.join("structures", "1em7.pdb")
kNumMaxIterations = 200
kNumSteps = 20_000
kRestartSimulation = False
kFixProtein = False

# CUDA is recommended for protein simulations (example output is from a RTX 4090)
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
simulator = Simulator(output_folder=f"{kSimulationName}/",
                      device_ml=device,
                      device_mm=device)
simulation = simulator(pdb_input=kPdbInput, 
                       max_iterations=kNumMaxIterations, 
                       restart=kRestartSimulation, 
                       fix=kFixProtein)
simulation.step(kNumSteps)
```
